3.1. Створіть нову колонку в даних `Rating Category`, яка визначає категорію рейтингу: високий, середній чи низький за наступним правилом:
- якщо рейтинг вищий або рівний 8 - то він високий (`High`)
- якщо рейтинг від 6 до 8 не включаючи - то він середній (`Medium`)
- інакше - низький (`Low`).

```python
def Rating_category(val):
  if val >=8:
    return 'High'
  elif 8>val>6:
    return 'Medium'
  else:
    return 'Low'
```

```python
df['Rating Category'] = df.Rating.apply(Rating_category)
df[['Rating Category', 'Rating']]
```

3.2. Виведіть, який відсоток даних становить кожне зі значень категорій рейтингу.

```python
total_value = df[['Rating Category']].count()
high_per = df[df['Rating Category'] == 'High'].count() / total_value * 100
medium_per = df[df['Rating Category'] == 'Medium'].count() / total_value * 100
low_per = df[df['Rating Category'] == 'Low'].count() / total_value * 100
print('High =', high_per['Rating Category'], 'Medium =', medium_per['Rating Category'], 'Low = ', low_per['Rating Category'])
```
